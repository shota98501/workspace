```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('../Datasets/sales_superstore_raw.csv')
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed').dt.strftime('%Y-%m-%d')
df
```

```python
print(df.dtypes)
```

```python
df['Ship Mode'] = df['Ship Mode'].fillna('Unknown')
df
```

```python
df.isnull().sum()
```

```python
df['Segment'].unique()
```

```python
df['Segment'] = df['Segment'].str.title()
df['Segment'].unique()
```

```python
df['Segment'] = df['Segment'].replace(['Consumr','Counsumer','  Consumer'], 'Consumer')
df['Segment'].unique()
```

```python
df['Segment'] = df['Segment'].replace(['Home-Office','Homeoffice'], 'Home Office')
df['Segment'].unique()
```

```python
df['Segment'] = df['Segment'].replace(['Corporate  ','Corp.','Corperate'], 'Corporate')
df['Segment'].unique()
```

```python
df['Country/Region'].unique()
```

```python
df['City'].unique()
```

```python
df
```

```python
df['Category'].unique()
```

```python
df['Category'] = df['Category'].replace('  Furniture   ', 'Furniture')
df['Category'].unique()
```

```python
df['Sub-Category'].unique()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Sales'] = df['Sales'].fillna(df['Sales'].median())
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Quantity'] = df['Quantity'].astype(int)
```

```python
df['Profit'] = df.groupby('Category')['Profit']\
    .transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df
```

```python
df = df.drop('Product Name', axis=1)
df
```

```python
df[['Sales','Discount','Profit']] = df[['Sales','Discount','Profit']].round().astype(int)
df
```

```python
df.to_csv('CLEAN_Superstore_Sales.csv', index=False)
```
