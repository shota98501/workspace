```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../Datasets/Netflix1.csv')
data
```

```python
data.isnull().sum()
```

```python
type = data['type'].value_counts().reset_index()
type.columns = ['type', 'count']
type
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x =type['count'], y = type['type'] , palette = 'rocket')
plt.xticks(rotation= 90)
plt.show
```

```python
duration = data['duration'].value_counts().reset_index()
duration.columns = ['duration', 'count']
duration
```

```python
import re
```

```python
season = data['duration'].str.extract(r'(\d+\s*) season' , flags = re.IGNORECASE).astype(object)

season = season.value_counts().reset_index()

season.columns = ['num_season' , 'count']

season
```

```python
minutes = data['duration'].str.extract(r'(\d+\s*) min' , flags = re.IGNORECASE).astype(object)

minutes = minutes.value_counts().reset_index()

minutes.columns = ['minutes' , 'count']

minutes
```

```python
plt.figure(figsize = (25,19))

plt.subplot(2,1,1)
sns.barplot(x = season['num_season'] , y = season['count'] ,  palette = 'rocket')
plt.xticks(rotation = 90)
plt.title('Seasons')

plt.subplot(2,1,2)
sns.barplot(x = minutes['minutes'] , y = minutes['count'] , data = minutes[:100] , palette = 'rocket')
plt.xticks(rotation = 90)
plt.title('Minutes')

plt.show()
```
