```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
clean_data = pd.read_csv('../Datasets/personality_datasert.csv')
clean_data
```

```python
data = pd.read_csv('../Datasets/personality_dataset.csv')
data
```

```python
data.isnull().sum()
```

```python
mean_data = data['Time_spent_Alone'].mean()
data['Time_spent_Alone'] = data['Time_spent_Alone'].fillna(mean_data)
data
```

```python
data['Time_spent_Alone'] = data['Time_spent_Alone'].round()
data
```

```python
data.isnull().sum()
```

```python
mode_fear = data['Stage_fear'].mode().iat[0]
data['Stage_fear'] = data['Stage_fear'].fillna(mode_fear)
data
```

```python
data.isnull().sum()
```

```python
mean_event = data['Social_event_attendance'].mean()
data['Social_event_attendance'] = data['Social_event_attendance'].fillna(mean_event)
data
```

```python
mean_outside = data['Going_outside'].mean()
data['Going_outside'] = data['Going_outside'].fillna(mean_outside)
data
```

```python
data.isnull().sum()
```

```python
mean_friends = data['Friends_circle_size'].mean()
data['Friends_circle_size'] = data['Friends_circle_size'].fillna(mean_friends)
data
```

```python
mean_post = data['Post_frequency'].mean()
data['Post_frequency'] = data['Post_frequency'].fillna(mean_post)
data
```

```python
mean_drained = data['Drained_after_socializing'].mode().iat[0]
data['Drained_after_socializing'] = data['Drained_after_socializing'].fillna(mean_drained)
data
```

```python
data.isnull().sum()
```

```python
data.groupby('Personality').sum().round()[['Time_spent_Alone' , 'Social_event_attendance' , 'Going_outside' , 'Friends_circle_size' , 'Post_frequency']]
```

```python
time_spent_data = data.groupby('Personality').sum().round()['Time_spent_Alone']
time_spent_data
```

```python
time_spent_data = pd.DataFrame(time_spent_data)
```

TIme Spent Alone

```python
plt.figure(figsize=(15, 9))
sns.barplot(data = time_spent_data, x = 'Time_spent_Alone', y= 'Personality' , palette='rocket')
plt.xticks(rotation= 90)
plt.show
```
