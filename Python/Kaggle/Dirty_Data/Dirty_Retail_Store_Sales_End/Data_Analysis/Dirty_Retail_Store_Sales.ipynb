```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =pd.read_csv('../Datasets/retail_store_sales.csv')
data
```

```python
data.isnull().sum()
```

```python
data['Transaction ID'] = data['Transaction ID'].str.replace('TXN_', '')
data
```

```python
data['Customer ID'] = data['Customer ID'].str.replace('CUST_', '')
data
```

```python
data['Item'] = data['Item'].fillna('Unknown')
data
```

```python
data.isnull().sum()
```

```python
data['Discount Applied'] = data['Discount Applied'].fillna('Unknown')
data
```

```python
data.isnull().sum()
```

```python
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce')
data['Price Per Unit'] = pd.to_numeric(data['Price Per Unit'], errors='coerce')
data['Total Spent'] = pd.to_numeric(data['Total Spent'], errors='coerce')


data['Price Per Unit'] = data['Price Per Unit'].fillna((data['Total Spent'] / data['Quantity']).round(2))
data['Total Spent'] = data['Total Spent'].fillna((data['Price Per Unit'] * data['Quantity']).round(2))
data['Quantity'] = data['Quantity'].fillna((data['Total Spent'] / data['Price Per Unit']).round(2))

data['Price Per Unit'] = data['Price Per Unit'].fillna('Unknown')
data['Total Spent'] = data['Total Spent'].fillna('Unknown')
data['Quantity'] = data['Quantity'].fillna('Unknown')

data
```
