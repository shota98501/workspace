```python
from dateutil import parser
```

```python
def parse_date(date_str):
    try:
        return pd.to_datetime(date_str, format="%Y-%m-%d")
    except:
        try:
            return parser.parse(date_str, dayfirst=True)
        except:
            return pd.NaT

data['checkin_date'] = data['checkin_date'].apply(parse_date)
data['checkout_date'] = data['checkout_date'].apply(parse_date)
```

```python
data[['checkin_date','checkout_date']] 
```

```python
data['nights'] = (data['checkout_date'] - data['checkin_date']).dt.days
data
```
