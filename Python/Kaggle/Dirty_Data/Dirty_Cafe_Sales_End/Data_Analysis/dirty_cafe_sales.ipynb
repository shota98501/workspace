```python
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce')
data['Price Per Unit'] = pd.to_numeric(data['Price Per Unit'], errors='coerce')
data['Total Spent'] = pd.to_numeric(data['Total Spent'], errors='coerce')


data['Price Per Unit'] = data['Price Per Unit'].fillna((data['Total Spent'] / data['Quantity']).round(2))
data['Total Spent'] = data['Total Spent'].fillna((data['Price Per Unit'] * data['Quantity']).round(2))
data['Quantity'] = data['Quantity'].fillna((data['Total Spent'] / data['Price Per Unit']).round(2))

data['Price Per Unit'] = data['Price Per Unit'].fillna('Unknown')
data['Total Spent'] = data['Total Spent'].fillna('Unknown')
data['Quantity'] = data['Quantity'].fillna('Unknown')

data.head()
```

```python
data.isnull().sum()
```

```python
data['Item'] = data['Item'].fillna("Uknown")
data['Item'] = data['Item'].str.strip().str.title()
data['Item'].unique()
```

```python
data['Payment Method'] = data['Payment Method'].fillna("Uknown")
data['Payment Method'] = data['Payment Method'].str.strip().str.title()
data['Payment Method'].unique()
```

```python
data['Location'] = data['Location'].fillna("Uknown")
data['Location'] = data['Location'].str.strip().str.title()
data['Location'].unique()
```

```python
data.isnull().sum()
```

```python
data["Transaction Date"] = pd.to_datetime(data['Transaction Date'], errors='coerce')

mode_date = data["Transaction Date"].mode().iat[0]
data['Transaction Date'] = data['Transaction Date'].fillna(mode_date)

data['Transaction Date'].head(12)
```
