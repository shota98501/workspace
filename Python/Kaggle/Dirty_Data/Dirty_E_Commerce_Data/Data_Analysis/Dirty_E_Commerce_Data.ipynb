```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
directory = '../Datasets'

dfs = []

for file in os.listdir(directory):
    if file.endswith('.csv'):
        file_path = os.path.join(directory, file)
        category = file.split('-')[-2]

        df = pd.read_csv(file_path)
        df['category'] = category
        dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)

combined_df
```

```python
combined_df.info()
```

```python
missing_data = (combined_df.isnull().sum() / len(combined_df)) * 100
print(f'Percentage of missing data per column: \n{missing_data}')
```

```python
df = combined_df.copy()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['goods-title'] = df['goods-title-link--jump'].fillna('') + df['goods-title-link'].fillna('')
df
```

```python
df['goods-title'].isnull().sum()
```

```python
df = df.drop(['goods-title-link--jump', 'goods-title-link'], axis=1)
df
```

```python
df = df.drop(['goods-title-link--jump href', 'product-locatelabels-img src', 'blackfridaybelts-bg src', 'blackfridaybelts-content'], axis=1)
df
```

```python
df.isnull().sum()
```

```python
df['price'] = df['price'].replace('[\$,]' , '', regex=True).astype(float)
df
```

```python
category_mean_price = df.groupby('category')['price'].median()

df['price'] = df['price'].fillna(df['category'].map(category_mean_price))

df['discount'] = df['discount'].replace('[%]', '', regex=True).fillna(0).astype(int)
```

```python
df.isnull().sum()
```

```python
def transform_sp(selling_prop):
    if pd.isna(selling_prop):
        return 0
    
    selling_prop = str(selling_prop).lower().replace('+', '').replace(' sold recently', '')

    return int(float(selling_prop.replace('k', '')) * 1000) if 'k' in selling_prop else int(selling_prop)

df['selling_proposition'] = df['selling_proposition'].apply(transform_sp)
```

```python
df
```

```python
df['color-count'] = df['color-count'].fillna(0).astype(int)
df
```

```python
df['rank-title'] = df['rank-title'].replace('#' , '', regex=True).replace(r' Best Seller[s]?','',regex=True).fillna(0).astype(int)

df = df.drop(['rank-sub'], axis=1)
df
```

```python
missing_data = (df.isnull().sum() / len(df)) * 100
print(f'Percentage of missing data per column: \n{missing_data}')
```

```python
df.to_csv('CLEAN_ecommerce_data.csv', index=False)
```
